## Desafio: Sistema de Notas de Alunos


Crie um programa que gerencia as notas de alunos de uma turma. O programa deve permitir:

- Adicionar alunos e suas respectivas notas (várias notas por aluno).
- Calcular a média de cada aluno e exibir a lista de alunos com suas médias.
- Exibir o aluno com a maior média.
- Permitir que o usuário continue inserindo alunos e notas até decidir parar.

---

**Exemplo de Execução:**

_Inputs do usuário_

> Deseja adicionar um aluno? (S/N): S
>
> Digite o nome do aluno: João
>
> Digite as notas de João (separadas por espaço): 7.5 8.0 6.5
>
> Deseja adicionar um aluno? (S/N): S
>
> Digite o nome do aluno: Maria
>
> Digite as notas de Maria (separadas por espaço): 9.0 8.5 10.0
>
> Deseja adicionar um aluno? (S/N): N

_Outputs do programa_

> Médias dos alunos:
>
> João: 7.33
>
> Maria: 9.17
>
> A maior média é de Maria: 9.17
