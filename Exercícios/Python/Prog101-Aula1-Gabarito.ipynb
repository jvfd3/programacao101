## 1. Vamos montar uma calculadora!


Escreva um código que recebe (com a função input) dois números do usuário e guarde nas variáveis "numero1" e "numero2". Em seguida, calcule a soma desses valores e guarde na variável "soma". Por fim, imprima a mensagem "A soma dos valores é:", seguido do valor da soma.

## 4. Divisão de grupos por cores


Montaremos um sistema de divisão de pessoas em dois grupos de acordo com as cores que as pessoas escolherem. O sistema deve receber duas cores como entrada e, de acordo com a tabela abaixo, deve imprimir o grupo resultante. Caso as cores escolhidas não estejam na tabela, o sistema deve imprimir "Grupo inválido".

| Cor 1   | Cor 2    | Grupo Resultante |
| ------- | -------- | ---------------- |
| Amarelo | Azul     | Grupo 1          |
| Amarelo | Vermelho | Grupo 2          |
| Rosa    | Verde    | Grupo 1          |
| Rosa    | Roxo     | Grupo 2          |


### 4. Gabarito


```python
cor1 = input('Digite uma cor: ')
cor2 = input('Digite uma cor: ')

"""
# primeira cor: amarelo
  segunda cor: azul -> grupo 1
  segunda cor: vermelho -> grupo 2
# primeira cor: rosa
  segunda cor: verde -> grupo 1
  segunda cor: roxo -> grupo 2
# se nenhuma das combinações acontecer -> grupo invalido
"""
if (cor1 == "amarelo"):
    if (cor2 == "azul"):
        print("grupo 1")
    elif (cor2 == "vermelho"):
        print("grupo 2")
    else:
        print("grupo invalido")
elif (cor1 == "rosa"):
    if (cor2 == "verde"):
        print("grupo 1")
    elif (cor2 == "roxo"):
        print("grupo 2")
    else:
        print("grupo invalido")
else:
    print("grupo invalido")
```

```python
"""
azul e amarelo
amarelo e azul
"""

cor1 = input('Digite uma cor: ')
cor2 = input('Digite uma cor: ')

combinacao1 = (cor1 == "amarelo" and cor2 == "azul")
combinacao2 = (cor1 == "azul" and cor2 == "amarelo")

if (combinacao1 or combinacao2):
    print("grupo 1")
else:
    print("grupo invalido")
```
