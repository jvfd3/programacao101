## 1. Vamos montar uma calculadora!


Escreva um código que recebe (com a função input) dois números do usuário e guarde nas variáveis "numero1" e "numero2". Em seguida, calcule a soma desses valores e guarde na variável "soma". Por fim, imprima a mensagem "A soma dos valores é:", seguido do valor da soma.


## 2. Vamos calcular um salário com base nas horas trabalhadas por dia.


Escreva um código que recebe (com a função input) dois números do usuário e guarde nas variáveis "numero1" e "numero2". "numero1" será o valor que o empregado recebe por horas trabalhadas e "numero2" será o numero de horas trabalhadas.

Você deve calcular o valor total do salário e colocar na variavel "salario". Por fim, imprima a mensagem: "Salário final: " e em seguida o valor do salário. Considere que o mês tem 30 dias.

exemplo:

valor da hora: 20, quantidade de horas por dia: 8, salario: 4800


## 3. Nesse exercício, vocês vão verificar se um número é par ou ímpar.


Para isso, leia um número inteiro com a função input e guarde na variável "num_int". Agora, verifique se o número digitado é par ou ímpar por meio do operador módulo '%'. Para isso, guarde o resultado de (num_int % 2) na variável "paridade".

**Exemplo: 5 % 2 resulta no número 1, 10 % 2 resulta no número 0**

Por fim, crie a variável eh_par, que vai receber a comparação (paridade == 0). Após isso, imprima a variável "eh_par". Se o resultado for True, então o número é par. Se for exibido False, o número é ímpar.


## 4. Divisão de grupos por cores


Montaremos um sistema de divisão de pessoas em dois grupos de acordo com as cores que as pessoas escolherem. O sistema deve receber duas cores como entrada e, de acordo com a tabela abaixo, deve imprimir o grupo resultante. Caso as cores escolhidas não estejam na tabela, o sistema deve imprimir "Grupo inválido".

| Cor 1   | Cor 2    | Grupo Resultante |
| ------- | -------- | ---------------- |
| Amarelo | Azul     | Grupo 1          |
| Amarelo | Vermelho | Grupo 2          |
| Rosa    | Verde    | Grupo 1          |
| Rosa    | Roxo     | Grupo 2          |
